```python
import os
import numpy as np
from pprint import pprint
from collections import defaultdict
```

```python
path = os.getcwd() + '/allPyPrograms_no_comments.txt'
dict_1 = defaultdict(int)
dict_2 = defaultdict(int)
```

```python
def add_to_dict(line):
    for i in range(len(line) - 1):
        if line[i : i + 2] in dict_1:
            dict_1[line[i : i + 2]] += 1
        else:
            dict_1[line[i : i + 2]] = 1
    
```

```python
file = open(path)
lines = file.readlines()
for line in lines:
    add_to_dict(line)
```

```python
pprint(dict_1)
```

```python
def recom(seen_seq, max_time):
    if max_time == 0:
        return "False"
    if seen_seq[0] == ' ':
        return seen_seq
    temp_dict = {k:v for k,v in dict_1.items() if k[0] == seen_seq}
    sorted_by_value = sorted(temp_dict.items(), key=lambda kv: kv[1], reverse = True)
    recom(sorted_by_value[0][0][1], max_time - 1)
```
